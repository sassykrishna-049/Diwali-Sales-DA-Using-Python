## Exploratory Data Analysis

### Gender 

```python
df_diwali.columns
```

```python
ax= sns.countplot(x='Gender',data= df_diwali)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax= sns.countplot(x='Gender',data= df_diwali)
```

```python
df_diwali.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
sales_gender = df_diwali.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x='Gender',y= 'Amount', data = sales_gender)
```

From the above graphs we could summarize that most of the buyers are female and even the purchasing power of females is higher than that of men.

## Age

```python
df_diwali.columns
```

```python
ax = sns.countplot(data=df_diwali,x= "Age Group",hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
## Total AMount vs Age group
sales_age_group = df_diwali.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x='Age Group',y= 'Amount', data = sales_age_group)
```

Most of the buyer are from 26-35 Age group Females

## State

```python
df_diwali.columns
```

```python
## Total number of orders from the top 10 states 
sales_state = df_diwali.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x= 'State',y = 'Orders')
```

```python
## Total amount /sales from top 10 states

## Total number of orders from the top 10 states 
sales_state = df_diwali.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x= 'State',y = 'Amount')
```

* From the above graphs, we can conclude that most of the orders are from the Uttar Pradesh, Maharashtra, Karnataka respectively *

## Marital Status

```python
ax = sns.countplot(data= df_diwali, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(7,3)})
for bars in ax.containers:
    ax.bar_label(bars)
    
    ## 0 stands for married and 1 stands for unmarried
```

```python
## Total AMount vs Age group
sales_state = df_diwali.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.set(rc = {'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y= 'Amount', data = sales_state,hue='Gender')
```

* From above Graphs we can see that most of the buyers are married(women) and have high purchasing power*

## Occupation


```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data= df_diwali,x = 'Occupation')


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
## Total Amount Spent in accordance with their Occupation
sales_occupation = df_diwali.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(x='Occupation',y= 'Amount', data = sales_occupation)

```

From above graphs we can see that most of the buyers are working in IT sector , Healthcare and Aviation

## Product Category 

```python
df_diwali.columns
```

```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data= df_diwali,x = 'Product_Category')


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
## Total Amount Spent in accordance with their Occupation
sales_Product_category = df_diwali.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.set(rc = {'figure.figsize':(17,5)})
sns.barplot(x='Product_Category',y= 'Amount', data = sales_Product_category)

```

*From the above graphs  we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_state = df_diwali.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc = {'figure.figsize':(17,5)})
sns.barplot(x='Product_ID',y= 'Orders', data = sales_state)
```

```python
# Top 10 most sold products

fig1,ax1 = plt.subplots(figsize = (12,7))
df_diwali.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending = False ).plot(kind= 'bar')
```
